# The dataset contains 1,871 records and 61 columns related to groundwater schemes in various districts. It includes details on state, district, and village locations, scheme types (e.g., tube wells), ownership, financial aspects, construction and maintenance costs, water distribution methods, and operational efficiency. It also covers technical specifications like bore depth, lifting devices, power sources, pumping duration, command area, and scheme utilization. Some fields have missing values, indicating incomplete data on certain aspects. The dataset can be used for groundwater resource management, scheme evaluation, and infrastructure planning.

# #finding mean,mode,median 

```python
if not df.select_dtypes(include=['number']).empty:
    mean_values = df.mean(numeric_only=True)
    median_values = df.median(numeric_only=True)
    mode_values = df.mode(numeric_only=True).iloc[0]
else:
    mean_values = median_values = mode_values = None
```

```python
df
```

```python
if mean_values is not None:
    df.fillna(mean_values, inplace=True)
```

```python
columns_to_plot = [
    "scheme_depth", 
    "nearest_scheme_distance", 
    "scheme_construction_cost", 
    "scheme_machinery_cost", 
    "scheme_maintainence_cost", 
    "ipc_total", 
    "ipu_total"
]
plt.figure(figsize=(12, 8))
for col in columns_to_plot:
    plt.plot(df.index, df[col], label=col, alpha=0.7)

plt.xlabel("Index")
plt.ylabel("Values")
plt.title("Line Plots for Selected Numerical Features")
plt.legend()
plt.grid()
plt.show()
```

```python
categorical_columns = ["scheme_type_name", "tubewell_type", "scheme_owner_name", "scheme_status_name"]
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
axes = axes.flatten()

for i, col in enumerate(categorical_columns):
    data_counts = df[col].value_counts()
    axes[i].pie(data_counts, labels=data_counts.index, autopct="%1.1f%%", startangle=140)
    axes[i].set_title(col.replace("_", " ").title())

plt.tight_layout()
plt.show()
```
