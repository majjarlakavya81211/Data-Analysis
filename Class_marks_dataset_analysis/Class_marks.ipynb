```python
s=s.astype("int64")
s
```

```python
s["Q1"] = s["Q1aM4"] + s["Q1bM6"]
s["Q2"] = s["Q2aM6"] + s["Q2bM4"]
s["Q3"] = s["Q3aM5"] + s["Q3bM5"]
s["Q4"] = s["Q4aM3"] + s["Q4bM7"]
s["Q5"] = s["Q5M10"]
s["Q6"] = s["Q6aM4"] + s["Q6bM6"]
s.drop(["Q1aM4", "Q1bM6","Q2aM6","Q2bM4" ,"Q3aM5" , "Q3bM5", "Q4aM3" ,"Q4bM7","Q5M10","Q6aM4","Q6bM6"], axis=1, inplace=True)
```

```python
s
```

```python
s["Percentage"]=s.Total/60 *100
s.fillna(0,inplace=True)
```

```python
s
```

```python
def assign_grade(percentage):
    if percentage >= 61:
        return 'A+'
    elif percentage >= 55:
        return 'A'
    elif percentage >= 45:
        return 'B+'
    elif percentage >= 30:
        return 'B'
    else:
        return 'F'

s['Grade'] = s['Percentage'].apply(assign_grade)

s
```

```python
grade_counts = s["Grade"].value_counts()
plt.figure(figsize=(8, 5))
grade_counts.plot(kind="bar", color="skyblue")
plt.title("Grade Distribution")
plt.xlabel("Grade")
plt.ylabel("Number of Students")
plt.xticks(rotation=0)
plt.show()
```
