```python
df['DV']=df['DV'].replace('MP',0)
```

```python
df['DV']=df['DV'].replace('A',0)
```

```python
df
```

```python
df=df.replace('MP',0)
df=df.replace('A',0)
df=df.replace('o',0)
df=df.replace('AB',0)
df=df.replace('I',1)
df=df.replace('II',11)
```

```python
df
```

```python
df[df['M-II'].isnull()]
```

```python
df.Total.count
```

```python
subject_columns = ['DV', 'M-II', 'PP', 'BEEE', 'FL', 'FIMS']
df[subject_columns] = df[subject_columns].apply(pd.to_numeric, errors='coerce')
df[subject_columns] = df[subject_columns].fillna(0)
df['Total'] = df[subject_columns].sum(axis=1)
df
```

```python
df["Percentage"]=df.Total/60 *100
df.fillna(0,inplace=True)
```

```python
df
```

```python
def assign_grade(percentage):
    if percentage >= 90:
        return 'A'
    elif percentage >= 80:
        return 'B'
    elif percentage >= 70:
        return 'C'
    elif percentage >= 60:
        return 'D'
    else:
        return 'F'

df['Grade'] = df['Percentage'].apply(assign_grade)

df
```

```python
def assign_grade(PP):
    if PP >= 18:
        return 'very good'
    elif PP>= 15:
        return 'good'
    elif PP >= 13:
        return 'average'
    else:
        return 'poor'

df['Performance'] = df['PP'].apply(assign_grade)

df
```

```python
df['DV'] = pd.to_numeric(df['DV'], errors='coerce')
def assign_grade(DV):
    if DV >= 18:
        return 'Very Good'
    elif DV >=15:
        return 'Good'
    elif DV >=13:
        return 'Average'
    else:
        return 'Poor'

df['performance_DV'] = df['DV'].apply(assign_grade)
df
```

# The histogram showa the distribution of total marks in a data set

```python
plt.figure(figsize=[8, 6])
plt.hist(df['Total'], color='blue', bins=10, edgecolor='black')
plt.title("Distribution of Total Marks")
plt.xlabel("Total Marks")
plt.ylabel("Number of Students")
plt.show()
```

```python
average_total_by_section = df.groupby('SECTION')['Total'].mean()
print("Average Total Marks by Section:")
print(average_total_by_section)
```

```python
plt.figure(figsize=[8, 6])
average_total_by_section.plot(kind='bar', color='black')
plt.title("Average Total Marks by Section")
plt.xlabel("Section")
plt.ylabel("Average Total Marks")
plt.xticks(rotation=45)
plt.show()
```

# DV vs M-II Marks by Section

```python
plt.figure(figsize=[8, 6])
for section in df['SECTION'].unique():
    section_data = df[df['SECTION'] == section]
    plt.scatter(section_data['DV'], section_data['M-II'], label=section, edgecolor='black')
plt.title("Scatter Plot: DV vs M-II Marks by Section")
plt.xlabel("DV Marks")
plt.ylabel("M-II Marks")
plt.legend(title="Sections")
plt.grid(True)
plt.show()
```
