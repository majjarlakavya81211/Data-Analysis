```python
columns_to_convert = ["DV", "M-II", "PP", "BEEE", "FL", "FIMS"]
for col in columns_to_convert: 
    df[col]=pd.to_numeric(df[col], errors='coerce')
df["Total"] = df["DV"] + df["M-II"] + df["PP"] + df["BEEE"] + df["FL"] + df["FIMS"]
df
```

```python
df["Percentage"]=df.Total/60 *100
df.fillna(0,inplace=True)
```

```python
def assign_grade(percentage):
    if percentage >= 90:
        return 'A'
    elif percentage >= 80:
        return 'B'
    elif percentage >= 70:
        return 'C'
    elif percentage >= 60:
        return 'D'
    else:
        return 'F'

df['Grade'] = df['Percentage'].apply(assign_grade)

df
```

```python
def assign_grade(PP):
    if PP >= 18:
        return 'very good'
    elif PP>= 15:
        return 'good'
    elif PP >= 13:
        return 'average'
    else:
        return 'poor'

df['Performance'] = df['PP'].apply(assign_grade)

df
```

```python
df['DV'] = pd.to_numeric(df['DV'], errors='coerce')
def assign_grade(DV):
    if DV >= 18:
        return 'Very Good'
    elif DV >=15:
        return 'Good'
    elif DV >=13:
        return 'Average'
    else:
        return 'Poor'

df['performance_DV'] = df['DV'].apply(assign_grade)
df
```

```python
import pandas as pd

# Load the Excel file
file_path = "MIDMARKS.xlsx"
df = pd.read_excel(file_path, sheet_name="SEM2 MID 1 - ALPHA")

# Define passing threshold
passing_marks = 10

# Convert subject columns to numeric (handling any non-numeric values)
subjects = ["DV", "M-II", "PP", "BEEE", "FL", "FIMS"]
for subject in subjects:
    df[subject] = pd.to_numeric(df[subject], errors="coerce").fillna(0)

# Convert scores to pass/fail (1 = Pass, 0 = Fail)
for subject in subjects:
    df[subject] = (df[subject] >= passing_marks).astype(int)

# Group by section and calculate pass probabilities
section_probabilities = df.groupby("SECTION")[subjects].mean() * 100  # Convert to percentage

# Create a final DataFrame with pass and fail probabilities
final_probabilities = {}
for section in section_probabilities.index:
    final_probabilities[section] = {
        subject: {
            "Pass Probability (%)": round(section_probabilities.loc[section, subject], 2),
            "Fail Probability (%)": round(100 - section_probabilities.loc[section, subject], 2)
        }
        for subject in subjects
    }

# Convert to DataFrame for better readability
section_prob_df = pd.concat(
    {k: pd.DataFrame(v).T for k, v in final_probabilities.items()}, axis=0
)

print(section_prob_df)
```

```python
import pandas as pd

for col in df.columns[2:]:
    df[col] = pd.to_numeric(df[col], errors='coerce')
passing_marks = 10
subject_section_pass = df.groupby('SECTION').apply(
    lambda df: (df.iloc[:, 2:] >= passing_marks).sum()
)
subject_section_fail = df.groupby('SECTION').apply(
    lambda df: (df.iloc[:, 2:] < passing_marks).sum()
)
total_students_per_section = df.groupby('SECTION').size()
prob_pass = subject_section_pass.div(total_students_per_section, axis=0)
prob_fail = subject_section_fail.div(total_students_per_section, axis=0)
print("Number of students passing each subject in each section:")
print(subject_section_pass)
print("\nNumber of students failing each subject in each section:")
print(subject_section_fail)
print("\nProbability of students passing each subject in each section:")
print(prob_pass)
print("\nProbability of students failing each subject in each section:")
print(prob_fail)
```
